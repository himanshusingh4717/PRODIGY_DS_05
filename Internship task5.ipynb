```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/user/Downloads/RTA Dataset.zip")
```

```python
df.head()
```

```python
req_columns = ['Time','Day_of_week','Age_band_of_driver','Sex_of_driver','Driving_experience','Type_of_vehicle','Service_year_of_vehicle','Accident_severity','Light_conditions','Weather_conditions']
```

```python
data = df[req_columns]
```

```python
data = data.dropna(how='any')
```

```python
data.head()
```

```python
data['Day_of_week'].value_counts().plot(kind='bar')
plt.xlabel('Day of Week')
plt.ylabel('No. of accidents')
plt.title('Distibution of accidents on different days of a week')
```

```python
data['Age_band_of_driver'].value_counts().plot(kind='bar')
plt.xlabel('Age of the driver')
plt.ylabel('No. of accidents')
plt.title('Relation between number of accidents and experience of the driver')
```

```python
pd.crosstab(data['Driving_experience'],data['Accident_severity'])
```

```python
pd.crosstab(data['Driving_experience'],data['Accident_severity']).plot(kind='bar')
```
